```python
import matplotlib.pyplot as plt
from math import log, exp

Na = 300
Nb = 200
q_total = 100
qa_list = range(q_total)
omega_total = []

def factor(N):
    N = int(N)
    total = 1
    if N<0:
        print('Error, N must be greater than 0')
    elif N>1: 
        for i in range(2,N+1):
            total *= i
    return total

def Multiplicity(N,n):
    return exp(log(factor(N)) - log(factor(n)) - log(factor(N-n)))

for qa in qa_list:
    qb = q_total - qa
    omega_a = Multiplicity(Na, qa)
    omega_b = Multiplicity(Nb, qb)
    omega_total.append(omega_a*omega_b)
plt.plot(qa_list, omega_total)  
plt.xlabel('$q_a$')
plt.ylabel('$\Omega$')
plt.show()
```
